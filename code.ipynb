```python
import pandas as pd

# Membaca file Excel
df = pd.read_excel("dataset.xlsx")

# Menghitung jumlah baris
num_rows = df.shape[0]
print("Jumlah baris:", num_rows)
```

```python
# Cek jumlah missing values di setiap kolom
print(df.isnull().sum())

# Hapus baris dengan missing values (jika diperlukan)
df_cleaned = df.dropna()
print(df_cleaned.info())
```

Periksa Format Koordinat

```python
# Periksa beberapa nilai di kolom 'Org_lat_lon' dan 'Des_lat_lon'
print(df['Org_lat_lon'].head())
print(df['Des_lat_lon'].head())
```

Pisahkan Kolom Koordinat dengan Memperbaiki Format

```python
# Pisahkan kolom 'Org_lat_lon' dan 'Des_lat_lon' menjadi latitude dan longitude
df[['Org_lat', 'Org_lon']] = df['Org_lat_lon'].str.split(',', expand=True).astype(float)
df[['Des_lat', 'Des_lon']] = df['Des_lat_lon'].str.split(',', expand=True).astype(float)

# Tampilkan hasil pemisahan kolom koordinat
print(df[['Org_lat', 'Org_lon', 'Des_lat', 'Des_lon']].head())
```

Pastikan Tidak Ada Nilai NaN

```python
# Periksa apakah ada nilai yang hilang setelah pemisahan koordinat
print(df[['Org_lat', 'Org_lon', 'Des_lat', 'Des_lon']].isnull().sum())
```

Memperbaiki Tipe Data

```python
# Pastikan tipe data kolom koordinat adalah float
df['Org_lat'] = df['Org_lat'].astype(float)
df['Org_lon'] = df['Org_lon'].astype(float)
df['Des_lat'] = df['Des_lat'].astype(float)
df['Des_lon'] = df['Des_lon'].astype(float)

# Periksa tipe data kolom-kolom tersebut
print(df[['Org_lat', 'Org_lon', 'Des_lat', 'Des_lon']].dtypes)
```

Menangani Koordinat yang Tidak Valid

```python
# Hapus baris dengan koordinat yang tidak valid (misalnya NaN atau koordinat yang tidak masuk akal)
df_cleaned = df.dropna(subset=['Org_lat', 'Org_lon', 'Des_lat', 'Des_lon'])

# Tampilkan informasi dataset yang sudah dibersihkan
print(df_cleaned.info())
```

Menghitung Jarak Antar Lokasi

```python
from geopy.distance import geodesic

# Fungsi untuk menghitung jarak antara dua titik
def calculate_distance(row):
    origin = (row['Org_lat'], row['Org_lon'])
    destination = (row['Des_lat'], row['Des_lon'])
    return geodesic(origin, destination).km

# Terapkan fungsi calculate_distance ke dataset
df['Distance'] = df.apply(calculate_distance, axis=1)

# Tampilkan beberapa baris dengan kolom 'Distance'
print(df[['Org_lat', 'Org_lon', 'Des_lat', 'Des_lon', 'Distance']].head())
```

Persiapan Data untuk CVRP

```python
# Misalkan kita memiliki kapasitas kendaraan dan jumlah barang
vehicle_capacity = 100  # Contoh kapasitas kendaraan
df['Load'] = df['TRANSPORTATION_DISTANCE_IN_KM']  # Asumsi bahwa load berhubungan dengan jarak (Anda bisa menyesuaikan sesuai dataset)

# Cek apakah kapasitas kendaraan cukup untuk mengirimkan barang
df['Can_Ship'] = df['Load'] <= vehicle_capacity

# Tampilkan beberapa baris dengan status pengiriman
print(df[['vehicle_no', 'Load', 'Can_Ship']].head())
```

Implementasi Algoritma Evolusi (Genetic Algorithm)

```python
from deap import base, creator, tools, algorithms
import random

# Definisikan masalah sebagai minimisasi jarak
creator.create("FitnessMin", base.Fitness, weights=(-1.0,))
creator.create("Individual", list, fitness=creator.FitnessMin)

# Fungsi untuk menghasilkan individu (solusi rute)
def create_individual():
    return [random.randint(0, len(df)-1) for _ in range(len(df))]

# Fungsi untuk menghitung biaya (jarak total)
def evaluate(individual):
    total_distance = 0
    for i in range(1, len(individual)):
        total_distance += df.iloc[individual[i-1]]['Distance'] + df.iloc[individual[i]]['Distance']
    return total_distance,

# Setup DEAP
toolbox = base.Toolbox()
toolbox.register("individual", tools.initIterate, creator.Individual, create_individual)
toolbox.register("population", tools.initRepeat, list, toolbox.individual)
toolbox.register("mate", tools.cxTwoPoint)
toolbox.register("mutate", tools.mutShuffleIndexes, indpb=0.1)
toolbox.register("select", tools.selTournament, tournsize=3)
toolbox.register("evaluate", evaluate)

# Inisialisasi populasi
population = toolbox.population(n=50)

# Algoritma evolusi
algorithms.eaSimple(population, toolbox, cxpb=0.7, mutpb=0.2, ngen=40, verbose=True)
```

Evaluasi Hasil

```python
# Tampilkan individu terbaik dari hasil evolusi
best_individual = tools.selBest(population, 1)[0]
print(f"Best route: {best_individual}")

# Hitung jarak total dari rute terbaik
total_distance = evaluate(best_individual)[0]
print(f"Total distance for best route: {total_distance} km")
```
